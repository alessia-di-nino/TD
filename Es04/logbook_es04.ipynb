Abbiamo preso $R_1 = 1k \Omega$ e $R_2 = 100k \Omega$, per avere un guadagno di 100. Abbiamo quindi scelto $V_{in} = 0.025 V$ per evitare la saturazione dell'uscita dell'amplificatore (con 0.03V si visualizza la distorsione tipica della saturazione), che corrisponde dunque al valore più alto utile per ridurre il rumore della misura.
Abbiamo scelto $Offset = 0$ perché il segnale è più grande di un ordine di grandezza rispetto all'offset dell'amplificatore.

Notiamo, tuttavia, che il segnale in uscita presenta un leggero rumore di fondo, come si può vedere in figura 3:

Dunque, abbiamo costruito un partitore di tensione in ingresso per poter generare un segnale più ampio tramite W1, meno rumoroso, che venisse attenuato in modo consistente all'amplificazione da effettuare. Aumentando il numero di punti per decade aumenti la densità di informazione per unità logaritmica di frequenza,
migliorando la risoluzione nella zona critica del polo dominante e quindi la precisione e stabilità del fit.
