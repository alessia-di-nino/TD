## **Task 9**

Obiettivi: 
- Campionare un segnale a dente di sega a 1 kHz, variando i parametri di campionamento in maniera da evidenziare o sopprimere il fenomeno del leakage spettrale;
- verificare, approssimativamente, la correttezza dell’ampiezza dell’armonica in presenza di leakage;
- fare un grafico sperimentale che evidenzi il fenomeno del leakage e salvarlo su Leakage.pdf.

Per il primo obiettivo, campioniamo due scenari: uno senza leakage, e uno con leakage. L'osservazione da tenere a mente è che  il caso ottimale (senza leakage) si ottiene quando $f_0 = \frac{kf_s}{N}$ per k intero, mentre negli altri casi la trasformata si “sparpaglia” sulle componenti spettrali limitrofe.

Per il primo scenario, impostiamo i parametri di campionamento in modo da evitare il leakage. La condizione da rispettare è, dunque, che la frequenza del segnale $(f_0)$ sia un multiplo intero della risoluzione in frequenza dell'analisi ($f_{res} = \frac{f_s}{N}$). Il ragionamento dunque consiste nel fissare $f_0 = 10kHz$ e scegliere $f_s$ ed $N$ in modo da rispettare la condizione di campionamento ottimale. Per semplicità, fissiamo $N = 10000$ campioni, e scegliamo $f_s = 1MSa/s$, in modo che la risoluzione in frequenza sia $f_{res} = 100 Hz$, e dunque $f_0 = 10000 Hz = 100 \cdot f_{res}$, ottenendo condizione soddisfatta con k = 100. Procediamo dunque con l'acquisizione, ottenendo il seguente spettro:


<img src='FFT-Es02-1MSa-N=10000.png' Width=500>

Ora, per il secondo scenario (leakage), basta modificare leggermente uno dei due parametri, ad esempio $N$, portandolo a 8192 campioni, in modo che la risoluzione in frequenza diventi $f_{res} = \frac{1MHz}{8192} \simeq 122.07 Hz$, e dunque $f_0 = 10000 Hz \simeq 81.92 \cdot f_{res}$, non più intero, ottenendo così uno spettro nel quale ci aspettiamo essere leakage. Procediamo dunque con l'acquisizione, ottenendo il seguente spettro:

<img src='FFT-Es02-1MSa-N=8192.png' Width=500>
