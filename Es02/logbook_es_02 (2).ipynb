## **Task 5**

Obiettivo: Verificare la risposta in frequenza del filtro a campione, verificando che essa sia in accordo con quanto atteso, facendo pratica con lo script.

* 10 Hz: Sinusoide in ingresso con ampiezza a 6 V ed offset nullo, si registra un'attenuazione di circa 0.43 ed uno sfasamento apprezzabile; 

* 100 Hz: Con la stessa forma d'onda in ingresso (ma frequenza 100 Hz), osserviamo uno sfasamento minimo ed un'attenuazione di circa 0.68;

* 1 kHz: Si è mandata in ingresso la stessa forma d'onda con ampiezza di 3 V. Si oserva un guadagno approssimativamente di 0.67, anche in questo caso, come atteso, lo sfasamento è minimo;

* 10 kHz: Stessa forma d'onda del caso precedente (ovviamente con frequenza modificata), si misura un guadagno approssimativo di circa 0.32 ed un forte sfasamento.

Se lo sfasamento risulta in accordo qualitativo con le simulazioni (almeno in termini di valore assoluto), lo stesso non può essere affermato per il guadagno, per il quale, invece, convertendo in dB i valori ottenuti qualitativamente, si osserva un mismatch con le frequenze nella simulazione. Sovrapponendo la simulazione al fit effettuato nel tesk 7, il piccolo mismatch è effettivamente evidente.

Si noti inoltre l'indipendenza, attesa, dei risultati dall'ampiezza della forma d'onda.

### **Forma d'onda quadra**

In questo caso la forma d'onda in uscita cambia a seconda della frequenza della quadra in ingresso: a base frequenze rispetto alla prima frequenza di taglio l'azione del derivatore è dominante ed infatti la forma d'onda viene derivata (slope positiva del grafico del guadagno), mentre nella regione di guadagno massimo, tra le due frequenze di taglio, si ha una forma d'onda molto simile a quella in ingresso.

Infine per frequenze maggiori della frequenza di taglio dell'integratore, l'azione di quest'ultimo è dominante e quindi la forma d'onda in uscita risulta essere pari all'integrale di quella in ingresso.

L'attenuazione del segnale al variare della frequenza, invece, è minore rispetto al caso sinusoidale, grazie alle presenza delle armoniche di ordine superiore nella quadra, che si sommano all'armonica fondamentale.

### **Forma d'onda triangolare**

Per questa forma d'onda si registra una variazione della stessa meno sensibile ala variazione in frequenza, essenzialmente cambia solo a basse ed alte frequenze. In tali regimi si ritrovano le stesse caratteristiche della quadra, cioè derivazione a basse frequenze (vedo una pinna di squalo in uscita a 10 Hz) ed integrazione ad alte (archi di parabola concatenati a 10 kHz).

Obiettivo: Campionare un dente di sega a 10 kHz a frequenze di campionamento diverse avvicinandosi sempre più al limite di Nyquist, superandolo per poter osservare l'aliasing nella FT del segnale.

1. $f_c = 1$ MSa/S, si è ottenuto il seguente ASD:

Questa è una buona acquisizione che rispetta pienamente il criterio di Nyquist: tutte le armoniche sono ben visibili e l'altezza dei picchi rispetta qualitativamente l'andamento 1/n. È presente, però, una modulazione periodica dell'altezza dei picchi delle armoniche più alte dovuta al leakage spettrale, infatti l'acquisizione è stata fatta a $2^{14}$ punti, enfatizzando, quindi, il leakage; nel task successivo questa modulazione scomparirà del tutto, effettuando un'acquisizione più oculata.

2. $f_c = 38$ kSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_38kSa.jpg' Width=500>

Già a questa frequenza di campionamento l'aliasing diventa apprezzabile:

* Il picco centrale a 10 kHz coincide con l'armonica fondamentale del segnale.

* Invece i picchi ai suoi lati coincidono con le armoniche della replica negativa dello spettro traslato della frequenza di campionamento, sommate alle armoniche originali del segnale, questo spiega la loro irregolarità ed il loro totale discostamento dall'andamento 1/n atteso.

3. $f_c = 10$ kSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_10kSa.jpg' Width=500>

L'ASD di questo plot è di facile interpretazione, infatti, campionando il segnale con la sua stessa frequenza, ottengo un  punto per periodo, ovviamente sempre circa allo stesso livello, quindi attraverso l'oscilloscopio vedo un segnale in DC, la cui FT presenta un solo picco, proprio in corrispondenza dell'armonica nulla (relativa proprio all'offset del segnale).

4. $f_c = 10.5$ kSa/S, si è ottenuto il seguente ASD:

<img src='FFT_Es02_10-5kSa.jpg' Width=500>

Per questo campionamento l'interpretazione è leggermente più complicata, si osserva che:

* L'andamento dell'altezza dei picchi 1/n è circa riprodotto, questo significa che il segnale riprodotto dai campionamenti ha la stessa forma d'onda del segnale in ingresso (rampa di ddp);

* La spaziatura dei picchi è a 500 Hz invece che a 10 kHz, come visibile nel primo campionamento (quello accurato), ciò è dovuto al fatto che, campionando ad una frequenza così vicina a quella del segnale originale, si hanno pochi punti ogni periodo, quindi il segnale ottenuto dai campionamenti risulterà avere una frequenza minore (sarà rallentato); in più, in questo caso risulterà anche temporalmente invertito (ottengo una ramp down di potenziale). Quindi, l'ASD riportato è del tutto virtuale, essendo relativo ad un segnale apparente.


## **Task 9**

Obiettivi: 
- Campionare un segnale a dente di sega a 1 kHz, variando i parametri di campionamento in maniera da evidenziare o sopprimere il fenomeno del leakage spettrale;
- verificare, approssimativamente, la correttezza dell’ampiezza dell’armonica in presenza di leakage;
- fare un grafico sperimentale che evidenzi il fenomeno del leakage e salvarlo su Leakage.pdf.

Per il primo obiettivo, campioniamo due scenari: uno senza leakage, e uno con leakage. L'osservazione da tenere a mente è che  il caso ottimale (senza leakage) si ottiene quando $f_0 = \frac{kf_s}{N}$ per k intero, mentre negli altri casi la trasformata si “sparpaglia” sulle componenti spettrali limitrofe.

Per il primo scenario, impostiamo i parametri di campionamento in modo da evitare il leakage. La condizione da rispettare è, dunque, che la frequenza del segnale $(f_0)$ sia un multiplo intero della risoluzione in frequenza dell'analisi ($f_{res} = \frac{f_s}{N}$). Il ragionamento dunque consiste nel fissare $f_0 = 10kHz$ e scegliere $f_s$ ed $N$ in modo da rispettare la condizione di campionamento ottimale. Per semplicità, fissiamo $N = 10000$ campioni, e scegliamo $f_s = 1MSa/s$, in modo che la risoluzione in frequenza sia $f_{res} = 100 Hz$, e dunque $f_0 = 10000 Hz = 100 \cdot f_{res}$, ottenendo condizione soddisfatta con k = 100. Procediamo dunque con l'acquisizione, ottenendo il seguente spettro:


<img src='FFT-Es02-1MSa-N=10000.png' Width=500>

Ora, per il secondo scenario (leakage), basta modificare leggermente uno dei due parametri, ad esempio $N$, portandolo a 8192 campioni, in modo che la risoluzione in frequenza diventi $f_{res} = \frac{1MHz}{8192} \simeq 122.07 Hz$, e dunque $f_0 = 10000 Hz \simeq 81.92 \cdot f_{res}$, k non più intero, ottenendo così uno spettro nel quale ci aspettiamo essere leakage. Procediamo dunque con l'acquisizione, ottenendo il seguente spettro:

<img src='FFT-Es02-1MSa-N=8192.png' Width=500>

dove si intravede chiaramente il fenomeno del leakage, in quanto l'armonica a 10 kHz non è più un picco netto, ma si "sparge" sulle frequenze limitrofe (si può osservare dall'allargamento alla base dei primi picchi, e dalla modulazione tramite sinc nei picchi successivi). Per visualizzare meglio la differenza tra i due grafici, è utile effettuare uno "zoom" sull'armonica a 10 kHz, per guardare meglio la dispersione sui primi picchi, come mostrato di seguito:
